## Create class

```python
class HtmlObject():
    
    def __init__(self, month, day):
        self.month = month
        self.day = day
        
        url_front = 'https://www.epexspot.com/en/market-data/intradaycontinuous/intraday-table/'
        url_end = '/FR'
        self.url = url_front + '2018' + '-' + str(month).zfill(2) + '-' + str(day).zfill(2) + url_end
        self.soup = None
        self.dateCheck = None
        self.make_soup()
        
        self.rows = None
        self.prices = []
        self.extract_rows()
        self.extract_prices()

    def make_soup(self):
        r = requests.get(self.url)
        self.soup = BeautifulSoup(r.content, "html.parser")
        self.dateCheck = self.soup.find('span', {'class': 'date'})

    def extract_rows(self):
        table = self.soup.find('table')
        self.rows = table.find_all('tr')
        
        
        
    def extract_prices(self):
        def hours_in_day(month, day):
            if month==3 and day==25:
                return 24
            elif month==10 and day==28:
                return 25
            else:
                return 24
    
        hours = hours_in_day(self.month, self.day)
        for i in range(0, hours):
            try:
                target_value = float(self.rows[3+i*3].find_all('td')[14].text)
            except ValueError:
                target_value = self.rows[3+i*3].find_all('td')[14].text
            # self.prices = target_value
            self.prices.append(target_value)


days_in_month = {1: 31,
                 2: 28,
                 3: 31, 
                 4: 30, 
                 5: 31, 
                 6: 30, 
                 7: 31, 
                 8: 31, 
                 9: 30,
                 10: 31, 
                 11: 30,
                 12: 31}

full_price_list = []

for month in range(1, 13):
    for day in range(1, days_in_month[month]+1):
        print('Processing 2018-' + str(month).zfill(2) + '-' + str(day).zfill(2))
        object = HtmlObject(month, day)
        full_price_list = full_price_list + object.prices
print('Complete')
```

## Using pandas

```python
import csv

with open('prices.csv', 'w') as myfile:
     wr = csv.writer(myfile, quoting=csv.QUOTE_ALL)
     wr.writerow(full_price_list)   
    
```
